```python
def readFile(nameFile):
  with open(nameFile) as f:
    listText = []
    for line in f:
        if not line.isspace():
            listText.append(line)
            
  return listText
```

```python
texto1 = readFile('texto1.txt')
```

```python
texto1=(''.join(texto1))
```

```python
len(texto1)
```

```python
texto1
```

```python
palavras=(texto1.split())
```

```python
len(palavras)
```

```python
palavras
```

```python
contagem = dict()


for p in palavras:
    if p in contagem:
        contagem[p] += 1
    else:
        contagem[p] = 1
```

```python
contagem
```

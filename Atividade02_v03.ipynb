```python
import unidecode
import string
import re
```

```python
def limpar(tokens):

  t=tokens.copy()

  t2=[]

  reg_tag=re.compile(r'<[^>]+>')
  reg_url=re.compile(r'^https?:\/\/.*[\r\n]*')
  reg_num=re.compile(r'[0-9]+')

  #pacote de  stopwords
  stopwords=nltk.corpus.stopwords.words('portuguese')
  
  for i in range(len(t)):

    #todas para minuscula
    termo_minusculo = t[i].lower()

    #remove h√≠fen de termo composto
    termo_sem_hifen="".join(termo_minusculo.split("-"))

    #remove acentua√ß√£o, chars especial como cedilha
    termo_so_letra_normal= unidecode.unidecode(termo_sem_hifen) 

    #remove stop words e pontua√ß√£o
    if (termo_so_letra_normal not in stopwords) \
    and (termo_so_letra_normal not in string.punctuation) \
    and not(reg_tag.match(termo_so_letra_normal)) \
    and not(reg_url.match(termo_so_letra_normal)) \
    and not (reg_num.match(termo_so_letra_normal)):
    
      t2.append(termo_so_letra_normal)

  return t2

```

```python
tokens_limpos=limpar(tokens3)
print(tokens_limpos)
```

#Stemming e Lemmatiza√ß√£o

```python
from nltk.stem import RSLPStemmer
```

```python
nltk.download('rslp')
```

```python
def stemminzar(tokens):
  stemmer = RSLPStemmer()
  return [stemmer.stem(termo) for termo in tokens]
```

```python
tokens_stimados=(stemminzar(tokens_limpos))
```

```python
print(tokens_stimados)
```

#Compara√ß√µes

```python
def criar_grafico_palavras(lista_palavras):
   df=pd.DataFrame(lista_palavras, columns=['palavra'])
   df_sumario=df['palavra'].value_counts() \
      .rename_axis('palavra') \
      .reset_index(name='ocorrencias')  \
      .sort_values(['ocorrencias', 'palavra'], ascending=[False, True]) 
   df_sumario.plot(x="palavra", y="ocorrencias", kind="barh", color='red', figsize=(20, 180))
```

##P√≥s processamento

```python
criar_grafico_palavras(tokens_stimados)
```
